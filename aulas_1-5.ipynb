```python
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVR
```

```python
materias.append('NU_NOTA_TOTAL')
```

```python
dados_sem_nota_zero = dados_sem_nota_zero[materias].dropna()
```

```python
provas_entrada = ['NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']
prova_saida = 'NU_NOTA_MT'
```

```python
notas_entrada = dados_sem_nota_zero[provas_entrada]
notas_saida = dados_sem_nota_zero[prova_saida]
```

```python
x = notas_entrada
y = notas_saida
```

```python
seed = 4321

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.25, random_state=seed)
```

```python
modelo = LinearSVR()
modelo.fit(x_treino, y_treino)
```

```python
predicoes_matematica = modelo.predict(x_teste)
predicoes_matematica[:5]
```

```python
y_teste.head()
```

```python
# comparacao entre as notas previstas, no eixo x, com as nota reais, eixo y
# podemos ver que erramos na previsao. Quanto mais proximo de uma linha reta diagonal positiva, melhor
plt.figure(figsize=(10,8))
sns.scatterplot(x=predicoes_matematica, y=y_teste)
```

```python
# nosso modelo estÃ¡ enviesado
# quanto maior a nota, maior Ã© o erro da previsÃ£o
# existe uma correlacao entre notas reais e o erro. Quanto maior a nota mais eu erro

plt.figure(figsize=(10,8))
sns.scatterplot(x=y_teste, y=y_teste - predicoes_matematica)
plt.xlabel('Nota Real')

plt.ylabel('Nota Real - Nota Prevista')
```

```python
modelo_arvore = DecisionTreeRegressor(max_depth=3)
modelo_arvore.fit(x_treino, y_treino)
predicoes_matematica_arvore = modelo_arvore.predict(x_teste)
```

```python
mean_squared_error(y_teste, predicoes_matematica_arvore)
```

```python
from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold
import numpy as np

seed = 1232
np.random.seed(seed)

partes = KFold(n_splits=10, shuffle=True)

modelo_arvore = DecisionTreeRegressor(max_depth=2)
resultados = cross_validate(modelo_arvore, x, y, scoring="neg_mean_squared_error", cv=partes)
```

```python
# funcao para calcula limite inferior e superior
def calcula_mse(resultados):
    media = (resultados['test_score']*-1).mean()
    desvio_padrao = resultados['test_score'].std()
    lim_inferior = media - (2 * desvio_padrao)
    lim_superior = media + (2 * desvio_padrao)
    print(f'Intervalo de confianÃ§a: {lim_inferior} - {lim_superior}')
    
```

```python
calcula_mse(resultados)
```

```python
def regressor_arvore(nivel=1):
    seed = 1232
    np.random.seed(seed)
    partes = KFold(n_splits=10, shuffle=True)
    modelo_arvore = DecisionTreeRegressor(max_depth=nivel)
    resultados = cross_validate(modelo_arvore, x, y, scoring="neg_mean_squared_error", cv=partes, return_train_score=True)
    print(f"Treino = {(resultados['train_score']*-1).mean()} | Teste = {(resultados['test_score']*-1).mean()}")
```

```python
for i in range(1, 21):
    regressor_arvore(i)
```
