```
import pandas as pd
import matplotlib.pyplot as plt
```

```
fonte = 'https://github.com/alura-cursos/imersao-dados-2-2020/blob/master/MICRODADOS_ENEM_2019_SAMPLE_43278.csv?raw=true'
```

```
dados = pd.read_csv(fonte)
dados.head()
```

```
dados.shape
```

```
dados.describe()
```

```
dados.columns.values
```

```
dados['SG_UF_RESIDENCIA'].value_counts()
```

```
dados['NU_IDADE'].hist(bins=20, figsize=(10,8))
plt.title('Distribuicao de Idade')
```

```
dados['NU_INSCRICAO'].count()
```

```
# numeros relativos de candidato por idade
# quanto porcento do total de candidatos tem 50 anos?
round(dados['NU_IDADE'].value_counts()[50] / dados['NU_INSCRICAO'].count() * 100, 2)
```

```
# quero saber quem eh treineiro e a distribuicao por idade
dados[dados['IN_TREINEIRO'] == 1]['NU_IDADE'].hist()
dados[dados['IN_TREINEIRO'] == 0]['NU_IDADE'].hist()
```

```
# distribuicao de notas de redacao
dados['NU_NOTA_REDACAO'].hist(bins=20, figsize=(10,8))
```

```
# agora quero saber a distribuicao de notas por materia
materias = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_MT', 'NU_NOTA_LC', 'NU_NOTA_REDACAO']
dados[materias].describe()
```

```
# vamos visualizar os quantis das notas por materia
dados['NU_NOTA_LC'].plot.box(figsize=(10,8), grid=True)
```
