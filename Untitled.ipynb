```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("fivethirtyeight")
```

```python
names = ['grou_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(12, 10))

plt.subplot(221)
plt.bar(names, values)
plt.subplot(222)
plt.scatter(names, values)
plt.subplot(223)
plt.plot(names, values)
plt.subplot(224)
plt.title('Exemplo4')
plt.plot(names, values)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.subplot(1, 2, 1)
plt.barh([1, 2,3], [1, 2, 3])
plt.subplot(1, 2, 2)
plt.bar([1, 2,3], [1, 2, 3])
```
